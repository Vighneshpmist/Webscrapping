# importing libraries

```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.imdb.com/list/ls002615341/"
```

# website response

```python
res=requests.get(url)
```

```python
res
```

# parsing

```python
soup=BeautifulSoup(res.text)
```

```python
soup
```

```python
cartoon_names=soup.select(".lister-item-header a")
```

```python
cartoon_names
```

```python
cartoon_ratings=soup.select(".ipl-rating-star.small .ipl-rating-star__rating")
```

```python
cartoon_ratings
```

```python
cartoon_stars=soup.select(".text-muted a:nth-child(1)")
```

```python
cartoon_stars
```

```python
cartoon_genre=soup.select(".genre")
```

```python
cartoon_genre
```
